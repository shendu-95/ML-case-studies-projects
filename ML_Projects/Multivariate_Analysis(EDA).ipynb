## Three Possibilities while performing bivariate analysis: 
### a. Numerical- Numerical 
### b. Categorical-Categorical 
### c. Numerical- Categorical

```python
# Scatter plot(Numerical- Numerical)

sns.scatterplot(x= tips['total_bill'], y = tips['tip'])
```

```python
sns.scatterplot(x= tips['total_bill'], y = tips['tip'], hue = tips['sex']) # additional Dimension 'sex' as hue for detailed distinctions(multi-variate)
```

```python
sns.scatterplot(x= tips['total_bill'], y = tips['tip'], hue = tips['sex'], style = tips['smoker'])
```

```python
sns.scatterplot(x= tips['total_bill'], y = tips['tip'], hue = tips['sex'], style = tips['smoker'], size = tips['size'])
```

```python
# Barplot (Numerical - Categorical)

titanic.head()
```

```python
sns.barplot(x= titanic['Pclass'], y= titanic['Age'])
```

```python
# Boxplot (Numerical - Categorical)
sns.boxplot(x=titanic['Sex'],y=titanic['Age'])
```

```python
sns.set_theme()
sns.histplot(titanic[titanic['Survived']==0]['Age'], kde=True, stat = 'density', alpha = 0.3)
sns.histplot(titanic[titanic['Survived']==1]['Age'], kde=True, stat = 'density')
```

```python
sns.kdeplot(titanic[titanic['Survived']==0], x='Age')
sns.kdeplot(titanic[titanic['Survived']==1], x='Age')
```

```python
pd.crosstab(titanic['Pclass'], titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'], titanic['Survived']))
```
